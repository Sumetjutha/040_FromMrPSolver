```python
# import library
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
# create function
def dSdt(S, t, g, m, b):
    vx = S[0]
    vy = S[1]
    return [
        -b/m * np.sqrt(vx**2+vy**2) * vx, #dvx/dt
        -g - b/m * np.sqrt(vx**2+vy**2) * vy #dvy/dt
    ]
```

```python
t = np.linspace(0, 20, 100)
m = 80
g = 9.81
vt = -55
b =  m*g/vt**2
v0x, v0y = 50, 0
```

```python
sol = odeint(dSdt, y0=[v0x, v0y], t=t, args=(g, m, b))
```

```python
sol
```

```python
vx = sol.T[0]
vy = sol.T[1]
```

```python
plt.plot(t, vx)
```

```python
plt.plot(t, vy)
plt.axhline(vt, color='salmon')
```
