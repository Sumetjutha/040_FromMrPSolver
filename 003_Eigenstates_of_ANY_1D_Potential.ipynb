```python
# import library
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import eigh_tridiagonal
```

```python
N = 2000
dy = 1/N
y = np.linspace(0, 1, N+1)
```

```python
def mL2V(y):
    return 1000*np.exp(-(y-0.7)**2 / (2*0.05**2))
```

```python
V = mL2V(y)
```

```python
plt.plot(y, V)
```

```python
d = 1/dy**2 + mL2V(y)[1:-1]
e = -1/(2*dy**2) *np.ones(len(d)-1) 
```

```python
w, v = eigh_tridiagonal(d,e)
```

```python
v
```

```python
plt.plot(v.T[0])
plt.plot(v.T[1])
plt.plot(v.T[2])
plt.plot(v.T[3])
```

```python
plt.plot(v.T[0]**2)
plt.plot(v.T[1]**2)
plt.plot(v.T[2]**2)
plt.plot(v.T[3]**2)
```
